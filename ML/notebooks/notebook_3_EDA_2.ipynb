# detecting outliers


in the dataset we are working on, since input of wrong data (such as in height and temperature) is highly unlikely and will more represent the real world. 
Thus, we will detect them for normalizations

```python
numerical_df = df[numerical_columns]

#function for outlier
def detect_outliers(data):
    #an empty dictionary for printing the summary
    outlier_summary={}
    for col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5*IQR
        upper = Q3 + 1.5*IQR
        outliers= data[(data[col] < lower) | (data[col] > upper) ]
        outlier_summary[col] ={
            "Q1": Q1,
            "Q3": Q3,
            "IQR": IQR,
            "Lower_Bound": lower,
            "Upper_Bound": upper,
            "Outlier_Count": outliers.shape[0],
            "Outlier_Percentage": round((outliers.shape[0] / len(data))*100, 2)
        }
    return pd.DataFrame(outlier_summary).T

#calling function
outlier_report = detect_outliers(numerical_df)
print(outlier_report)
            
```

## ANOVA and Kruskal-Wallis tests

```python
numeric_features = df[numerical_columns]
numeric_features
```

```python
# importing library
from scipy.stats import f_oneway
from scipy.stats import kruskal
```

```python
#anova test
anova_results = {}

for col in numeric_features:
    groups = [
        df[df['strength'] == s][col]
        for s in sorted(df['strength'].unique())
    ]
    f_stat, p_val = f_oneway(*groups)
    anova_results[col] = p_val

anova_df = (
    pd.DataFrame.from_dict(anova_results, orient='index', columns=['anova_pvalue'])
    .sort_values('anova_pvalue')
)

anova_df
```

```python
#kruskal test
kruskal_results = {}

for col in numeric_features:
    groups = [
        df[df['strength'] == s][col]
        for s in sorted(df['strength'].unique())
    ]
    stat, p_val = kruskal(*groups)
    kruskal_results[col] = p_val

kruskal_df = (
    pd.DataFrame.from_dict(kruskal_results, orient='index', columns=['kruskal_pvalue'])
    .sort_values('kruskal_pvalue')
)

kruskal_df
```

# Training Baseline Models

We use models such as logistic regression and random forest, to form a baseline, which will work as a reference for the final model.

## logistic regression

Here, the target variable is "strength". Features sucha as "bucket" is unnecessary, and we have already done feature engineering to extract all the features from "password". So we drop that as well.

```python
# target
y = df['strength']

# features (drop non-numeric + target)
X = df.drop(columns=['password', 'bucket', 'strength'])
```

```python
#splitting data before scaling
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2, # this is to split 20% data for testing
    random_state=42,
    stratify=y   # this is for class imbalance
)
```

```python
#scaling the data (for logisitic)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
#training for logistic regrerssion
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(
    max_iter=1000,
    multi_class='multinomial',
    solver='lbfgs'
)

log_reg.fit(X_train_scaled, y_train)
```

```python
#evaluating the performance of the regression model
from sklearn.metrics import classification_report, confusion_matrix

y_pred = log_reg.predict(X_test_scaled)

print(classification_report(y_test, y_pred))
```

```python
#confusion matrix:
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

From the scores we see that for strength 0, the model rarely predicts strength 0 (low recall), but when done, its usually correct (high precision). This is due to low no. of samples and class imbalance, and thus we can conclude that minority class handling is required. Strength 1 and 2 have dominated here, so rather than trusting accuracy, we look at f1-score. From macro avg, we see a poor performance from class 0 and 3. We still see some good F1-score and above 50% accuracy, so we can conclude there is real predicitive information in the features. Also, we see that the non-linear interactions havent been handled properly, so we move to a tree based model.

rom previous, the class 0 recall, and hence the f1-score has improved. Since the recall value for class 2 and 3 have also improved, we can say that they have separated better. It is also shown from the confusion matrix, where the values in diagonal have gone up for 2 and 3. Also, the macro avg has also improved to 0.68, which shows that the model has handled the relation better than regression model.

# class balancing

```python
from sklearn.utils.class_weight import compute_class_weight

classes = np.unique(y_train)
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=y_train
)

class_weight_dict = dict(zip(classes, class_weights))
class_weight_dict
```

```python
sample_weights = y_train.map(class_weight_dict)

xgb_weighted = XGBClassifier(
    objective='multi:softprob',
    num_class=5,
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='mlogloss',
    random_state=42
)

xgb_weighted.fit(X_train, y_train, sample_weight=sample_weights)
```

```python
from sklearn.metrics import classification_report

y_pred = xgb_weighted.predict(X_test)
print(classification_report(y_test, y_pred))
```

Applying class-weighted XGBoost significantly improved recall for minority strength classes, particularly class 0 and class 4. However, this came at the cost of reduced precision and macro-averaged F1 score, indicating over-compensation for class imbalance. This trade-off highlights the need for controlled weighting and further hyperparameter tuning rather than aggressive imbalance correction.

# TUNING for better macro avg

We start again with a base model of the XGBoost. We define a parameter grid with different parameters we want to try, and use the gridsearch method to find the most optimized fit through the parameter grid.

```python
#defining base model
from xgboost import XGBClassifier

xgb_base = XGBClassifier(
    objective='multi:softprob',
    num_class=5,
    eval_metric='mlogloss',
    random_state=42,
    n_jobs=-1
)
```

```python
# defining the parameter grid
param_grid = {
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1],
    'n_estimators': [200, 400],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
#training through grid search
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(
    estimator=xgb_base,
    param_grid=param_grid,
    scoring='f1_macro',
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid.fit(X_train, y_train)
```

```python
best_xgb = grid.best_estimator_

y_pred = best_xgb.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

The macro avg, which is thee simple mean of all f1score, is only 0.69, which is lower. the target is to reach a minimum of 80%.
